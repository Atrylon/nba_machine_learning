```python
class Game:
    def __init__(self, date_game, game_start_time, visitor_team_name,visitor_pts, 
                 home_team_name, home_pts, box_score_text, overtimes, attendance, 
                 game_remarks):
        self.date_game = date_game
        self.game_start_time = game_start_time
        self.visitor_team_name = visitor_team_name
        self.visitor_pts = visitor_pts
        self.home_team_name = home_team_name
        self.home_pts = home_pts
        self.box_score_text = box_score_text
        self.overtimes = overtimes
        self.attendance = attendance
        self.game_remarks = game_remarks
        
    def get_list(self):
        return [self.date_game, self.game_start_time, self.visitor_team_name, 
                self.visitor_pts, self.home_team_name, self.home_pts, self.box_score_text,
                self.overtimes, self.attendance, self.game_remarks]
```

```python
def SaveInCsv(en_tete, data):
    
    for x in range(len(data)):
        ligne = []
        game = data[x]

        for attribut_label in attributs_labels:
            ok = False
            for label in game:
                if attribut_label == label:
                    ligne.append(game[label])
                    ok = True
                    break

            if ok == False:
                ligne.append("")
#                 print(None)


        with open('nba_games_data.csv', 'a', encoding="utf-8", newline='') as csvfile:
            spamwriter = csv.writer(csvfile, delimiter=';')
            if en_tete==True:
                spamwriter.writerow(attributs_labels)
                en_tete = False
            spamwriter.writerow(ligne)
    return en_tete
```
